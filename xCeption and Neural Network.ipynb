```python
from tensorflow.python import keras
from tensorflow.python.keras.applications.xception import Xception, decode_predictions, preprocess_input
from tensorflow.python.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = Xception(weights = "imagenet")
img_path = "Elephant4XCEPTION.jpeg"
img = image.load_img(img_path, target_size = (299,299))
img = np.asarray(img)
print(img.shape)
plt.imshow(img)
img = np.expand_dims(img, axis = 0)
img = preprocess_input(img)
preds = model.predict(img)
print("\nXception believes this is either a(n)",decode_predictions(preds, top= 3)[0][0][1], " a(n)",decode_predictions(preds, top= 3)[0][1][1], ",or a(n)",decode_predictions(preds, top= 3)[0][2][1])
```
